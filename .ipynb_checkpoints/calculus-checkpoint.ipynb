# Linear Function

$ y = mx + b$
<br>

*Example:*

  
*if x,y = 0 :*    

<center>  
    $y = 3x + 1$
<br>
    $0 = 3(0) + 1 = 1$

*Rate of change*:
 
$m = \frac{\text{change in y}}{\text{change in x}} = \frac{\bigtriangleup y}{\bigtriangleup x} = \frac{y_1 - y_2}{x_2 - x_1}$

````python
def slope(x1,x2,y1,y2):
    m = (y2-y1)/(x2-x1)
    return m
slope_one = slope(0,4,1,13)
slope_two = slope(5,-1,16,-2)
````

*Secant Line*: 
A line that intersects 2 points on a curve

---

$y = -(x^2) + 3x - 1$

if $x_1=0$ , $x_2=1$ :

<center>
$y_1 = 0 + 0 -1 = -1$
</center>
    
<center>
$y_2 = -1 + 3 -1 = 1$
</center>

then $x_1=0$ , $x_2=1$, $y_1=-1$, $y_2=1$:

<br>

<center>
$m = \frac{1-(-1)}{1}=2$
</center>

find y-intercept , with $(x_1,y_1)$ or $(x_2,y_2)$ or any point pnt the line:

<br>

<center>
$y_1=m*x_1 +b$    
</center>    

<center>
$-1 = 2(0) + b$
</center>

then $b=-1$

therefore $y=2x -1$


````python
import seaborn
seaborn.set(style='darkgrid')

def draw_secant(x_values):
    x = np.linspace(-20,30,100)
    y = -1*(x**2) + x*3 - 1
    y_values = [-1*(i**2) + i*3 - 1 for i in x_values]
    m = (y_values[1] - y_values[0])/(x_values[1] - x_values[0])
    c = y_values[0] - m*(x_values[0])
    y_secant = [m*i + c for i in x]
    plt.plot(x,y)
    # Add your code here.
    plt.plot(x,y_secant,c='green')
    plt.show()
````

````python
draw_secant([3,5])
````

````python
draw_secant([3,10])
````

````python
draw_secant([3,15])
````

At the end of the last mission, we fixed a first point on our curve, drew a secant line between that first point and a second point, and observed what happened when we moved the second point closer to the first point along the curve. The larger the interval between the 2 points on the x-axis, the more the steepness of the secant line diverged from the steepness of the curve. The closer the interval, the more the secant line started to match the steepness at the first point on the curve. 

---

For linear functions, the instantaneous rate of change at any point on the line is the same. 


For nonlinear functions, the instantaneous rate of change describes the steepness of the graph of the function at a specific point. 

### tangent

only intersects our function at one point


think of the tangent line as the secant line when both points are the same

# Limit

limit describes the value a function approaches when the input variable to the function approaches a specific value.


$m = \frac{f(x_2)-f(x_3)}{x_2 - x_1}$

*As* $x_2$ approaches 3, the slope between $x_1$ and $x_2$ approaches - 3 

$\lim_{x_2 -> 3} \frac{f(x_2) - f(x_1)}{(x_2 - x_1)}=-3$

## Defined Limits and Undefined Limits

### Undefined Limit

$\lim_{x_2 -> 3} \frac{f(x_2) - f(x_1)}{(x_2 - x_1)}=-3$

$f(x) = -(x^2) + 3x - 1$


*Fixed $x_1$ to 3 Then :*

<center>
$f(x_1) = -(3^2) + 3*3 - 1 = -1$
</center>
<br>
<center>
$\lim_{x_2 -> 3} \frac{f(x_2) - (-1)}{(x_2 - 3)}=-3$
</center>

*$x_2$ = 3 Then :*

<center>
$f(x_2) = -(3^2) + 3*3 - 1 = -1$
</center>
<br>
<center>
$\lim_{x_2 -> 3} \frac{-1 - (-1)}{(3 - 3)} = 0 $
</center>

###  Defined Limit

$\lim_{x_2 -> 3} \frac{f(x_2) - f(x_1)}{(x_2 - x_1)}=-3$

$f(x) = -(x^2) + 3x - 1$

<center>

## Property of Limits 

## First Property

$\lim_{x->a}[f(x) + g(x)] = \lim_{x->a}f(x) + \lim_{x->a}g(x)$
<br><br>
$\lim_{x->a}[f(x) - g(x)] = \lim_{x->a}f(x) - \lim_{x->a}g(x)$
<br><br>
*Example*
<br><br>
$ \lim_{x->1}[3x^2 + 3x -3]= \lim_{x->1}[3x^2] + \lim_{x->1}[3x] - \lim_{x->1}[3]$

## Second Property

 break up terms that are being multiplied by a constant

$\lim_{x->a}[cf(x)] = c \lim_{x->a}f(x)$

$\small \lim_{x->1}[3x^2] + \lim_{x->1}[3x] + \lim_{x->1}[3]$ = $ \small 3 \lim_{x->1}[x^2] + 3 \lim_{x->1}[x] - 3 \lim_{x->1}[x] $

````python
import sympy
x,y = sympy.symbols('x y')

limit_two = sympy.limit(
    3*(x**2) + 3*x -3,
    x,
    1
)

print(limit_two)
````

## Convert Undefined Limit to Defined Limit

$f(x) = -(x^2) + 3x - 1$
<br><br>
$\lim_{x_2 -> 3} \frac{f(x_2) - (-1)}{(x_2 - 3)}$<br><br>
$=\lim_{x_2 -> 3} \frac{-(x_2^2) + 3x_2 -1 - (-1)}{(x_2 - 3)}$
<br>
$=\lim_{x_2 -> 3} \frac{-x_2^2 + 3x_2 }{(x_2 - 3)}$
<br>
$=\lim_{x_2 -> 3} \frac{-x_2(x_2 - 3)}{(x_2 - 3)}$
<br>
$=\lim_{x_2 -> 3} -x_2$
<br>



````python
x2, y = sympy.symbols('x2 y')
y = (-x2**2 +3*x2 -1 + 1)/(x2 - 3)
limit_four = sympy.limit(y, x2, 3)
print(limit_four)
````

x = 0 and acted as a transition point from positive slope to negative slope. 

This is formally known as a **critical point**. At a function's critical point, 

> the slope is either 0 or undefined

Critical points are interesting to us in data science when they represent extreme values. Extreme values can be split into 2 categories -- minimum and maximum.

As we noticed at the beginning of the mission, a critical point can be a minimum value if the slope transitions from negative to positive at that point. 

When the slope transitions from positive to negative at a point, it can be a maximum value. If the slope doesn't transition at a critical point, like at  x = 0 for y = $x^3$
 in the last screen, then it can't be a minimum or a maximum point.
